## Create Embeddings and FAISS Vector Store 

1. Extract and combine the most relevant fields
2. (including media) Generate embeddings using Sentence Transformers
3. Create and save a FAISS vector store

### for the first file

```python
import json
import pandas as pd
from sentence_transformers import SentenceTransformer
from langchain.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings

# --- Step 1: Combine Relevant Fields (including media) ---
def combine_text_fields(row):
    parts = []
    for field in ['type', 'name', 'description', 'vision2030']:
        val = row.get(field)
        if pd.notnull(val):
            parts.append(str(val))
    for list_field in ['notable_sites', 'activities', 'sustainable_initiatives', 'main_projects', 'tourist_destinations']:
        items = row.get(list_field)
        if isinstance(items, list):
            for item in items:
                if isinstance(item, dict):
                    name = item.get('name') or item.get('project_name') or item.get('project')
                    desc = item.get('description')
                    if name or desc:
                        parts.append(f"{list_field[:-1].replace('_',' ').title()}: {name}. {desc}")
    if isinstance(row.get('values_and_beliefs'), dict):
        values = row['values_and_beliefs'].get('values', [])
        if isinstance(values, list):
            parts.append(f"Values: {', '.join(values)}")
    if isinstance(row.get('rituals'), list):
        parts.append("Rituals: " + ", ".join(row['rituals']))
    if isinstance(row.get('values'), list):
        parts.append("Values: " + ", ".join(row['values']))
    media = row.get('media')
    if isinstance(media, dict):
        for media_type in ['images', 'videos', 'audio']:
            urls = media.get(media_type, [])
            if isinstance(urls, list) and urls:
                parts.append(f"{media_type.title()}: {', '.join(urls)}")
    return " | ".join(parts)

texts = rimal_df.apply(combine_text_fields, axis=1).tolist()
metadatas = rimal_df.to_dict(orient="records")

# --- Step 2: Generate Embeddings and Save as LangChain FAISS Vector Store ---
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vectordb = FAISS.from_texts(texts, embedding_model, metadatas=metadatas)
vectordb.save_local("rimalai_faiss")

# Save the DataFrame as metadata for mapping results (optional)
rimal_df.to_csv('rimalai_metadata.csv', index=False)

print("Embeddings generated and FAISS vector store saved in 'rimalai_faiss' directory (LangChain format).")

```

# Set up audio

```python
from pvrecorder import PvRecorder
import wave
import struct

def record_audio(output_file="user_question.wav", duration=5):
    recorder = PvRecorder(frame_length=512)
    recorder.start()
    frames = []
    
    # Record for 'duration' seconds
    for _ in range(int(duration * 16000 / 512)):  # 16000Hz / 512 samples-per-frame
        pcm_frames = recorder.read()
        # Convert 16-bit integer samples to bytes
        bytes_frames = [struct.pack('h', frame) for frame in pcm_frames]
        frames.extend(bytes_frames)
    
    recorder.stop()
    with wave.open(output_file, 'wb') as f:
        f.setnchannels(1)
        f.setsampwidth(2)  # 16-bit = 2 bytes
        f.setframerate(16000)
        f.writeframes(b''.join(frames))
    recorder.delete()

# Test recording
record_audio(duration=25)
```

```python
import os
import openai
import langid
import tempfile
import ast
import re
import os
import speech_recognition as sr
from gtts import gTTS
from IPython.display import Audio, display
from langchain.chat_models import ChatOpenAI
from langchain.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate
from langchain.tools import Tool
from langchain.agents import initialize_agent, AgentType



# === 1. Load saved FAISS ===
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vectordb = FAISS.load_local(
    folder_path="rimalai_faiss",
    embeddings=embedding_model,
    allow_dangerous_deserialization=True
)

# === 2. Initialize LLM ===
openai.api_key = os.getenv("OPENAI_API_KEY")
client = openai
llm = ChatOpenAI(model_name="gpt-4o", temperature=0.3)

# === 3. Define prompt templates ===
custom_prompt = PromptTemplate(
    input_variables=["context", "question"],
    template="""
    You are a helpful assistant. Answer the following question using the provided context.
    If the question is in Arabic, answer in Arabic. If in English, answer in English.

    Context:
    {context}

    Question: {question}
    Answer:
    """
)

# === 4. QA Chains setup ===
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=vectordb.as_retriever(),
    return_source_documents=True,
    chain_type_kwargs={"prompt": custom_prompt}
)

def query_with_media(query: str):
    response = qa_chain(query)
    answer = response["result"]
    media_results = []
    for doc in response["source_documents"]:
        media = doc.metadata.get("media", {})
        if isinstance(media, str):
            try:
                media = ast.literal_eval(media)
            except Exception:
                media = {}
        video_link = media.get("videos")[0] if media.get("videos") else None
        image_link = media.get("images")[0] if media.get("images") else None
        audio_link = media.get("audio")[0] if media.get("audio") else None
        media_results.append({
            "title": doc.metadata.get("name"),
            "image": image_link,
            "video": video_link,
            "audio": audio_link
        })
    return {"answer": answer, "media": media_results}

# === Tools Functions ===

# @tool 1: QA with Media
db_tool = Tool(
    name="SaudiRimalRetrievalQAWithMedia",
    func=query_with_media,
    description="Answer questions about Saudi landmarks, Vision 2030, and culture, returning answer and media links."
)

# @tool 2: Arabic Culture Poem + Image
ARABIC_CULTURE_KEYWORDS = [
    "arab", "saudi", "bedouin", "najd", "hijaz", "mecca", "medina", "riyadh", "alula", "al-ula",
    "najdi", "ghazal", "nabati", "jeddah", "diriyah", "desert", "camels", "dates", "gahwa", "coffee",
    "palms", "arabian", "oasis", "sand", "tribe", "poet", "poetry", "heritage", "tradition",
    "قصيدة", "شعر", "عربي", "سعودي", "بدوي", "نجد", "الحجاز", "مكة", "المدينة", "الرياض", "العلا",
    "جدة", "الدرعية", "صحراء", "جمل", "تمر", "قهوة", "نخيل", "قبيلة", "شاعر", "تراث", "تقاليد"
]

def is_arabic_culture(prompt):
    prompt_lower = prompt.lower()
    return any(kw in prompt_lower for kw in ARABIC_CULTURE_KEYWORDS)

def generate_arabic_poem(prompt):
    system_prompt = (
        "أنت شاعر عربي محترف. "
        "اكتب قصيدة عربية أصيلة حول الموضوع الذي يطلبه المستخدم، بأسلوب تقليدي (مثل النبطي أو الغزل أو الحكمة). "
        "بعد القصيدة، قدم ترجمة إنجليزية دقيقة لها تحت عنوان 'Translation:'."
    )
    user_prompt = (
        f"اكتب قصيدة عربية حول: {prompt}\n"
        "ثم ترجم القصيدة إلى الإنجليزية تحت عنوان 'Translation:'."
    )
    response = client.ChatCompletion.create(
        model="gpt-4o",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ],
        max_tokens=400
    )
    return response.choices[0].message.content.strip()

def generate_image(prompt, n=1, size="1024x1024"):
    response = openai.Image.create(
        model="dall-e-3",
        prompt=prompt,
        n=n,
        size=size
    )
    return [img.url for img in response.data]

def arabic_culture_poem_and_image(prompt):
    if not is_arabic_culture(prompt):
        return {
            "message": (
                "⚠️ This tool is dedicated to Arabic culture and heritage. "
                "Please provide a prompt related to Arabic themes, history, or traditions."
            )
        }
    poem = generate_arabic_poem(prompt)
    images = generate_image(prompt)
    return {
        "poem": poem,
        "image_urls": images
    }

culture_tool = Tool(
    name="ArabicCulturePoemAndImage",
    func=arabic_culture_poem_and_image,
    description=(
        "Given a prompt related to Arabic culture, generates an original Arabic poem (with English translation) and a DALL-E image. "
        "If the prompt is not about Arabic culture, returns a polite message."
    )
)

# @tool 3: Text-to-Speech
def tts_tool_func(text: str) -> str:
    lang = langid.classify(text)[0]
    tts = gTTS(text, lang=lang if lang in ['en', 'ar'] else 'en')
    with tempfile.NamedTemporaryFile(delete=False, suffix=".mp3") as fp:
        tts.save(fp.name)
        return fp.name

tts_tool = Tool(
    name="GoogleTextToSpeech",
    func=tts_tool_func,
    description="Convert an answer in Arabic or English to speech (returns MP3 file path)."
)

# Voice Input Function (not a @tool)
def record_and_transcribe(duration=7, filename="user_question.wav"):
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        audio = r.record(source, duration=duration)
        with open(filename, "wb") as f:
            f.write(audio.get_wav_data())
    with open(filename, "rb") as f:
        transcript = openai.Audio.transcribe("whisper-1,f")

    return transcript["text"]

# === 6. Agent Setup ===
tools = [db_tool, culture_tool, tts_tool]
agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent=AgentType.OPENAI_FUNCTIONS,
    verbose=True
)

# === Voice Input Workflow ===
def record_audio(duration=7, filename="user_question.wav"):
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        audio = r.record(source, duration=duration)
        with open(filename, "wb") as f:
            f.write(audio.get_wav_data())
    return filename

def extract_audio_path(agent_result):
    """Extract .mp3 file path from agent result string."""
    match = re.search(r'([/\w\-.]+\.mp3)', agent_result)
    if match:
        path = match.group(1)
        if path.startswith("sandbox:"):
            path = path.replace("sandbox:", "")
        if not os.path.exists(path) and os.path.exists(os.path.basename(path)):
            path = os.path.basename(path)
        return path
    return None

def voice_query_agent_workflow(agent, duration=7):
    audio_file = record_audio(duration=duration)
    print("Transcribing...")
    with open(audio_file, "rb") as f:
        transcript = openai.Audio.transcribe("whisper-1", f)
    user_text = transcript["text"]
    print("\nYou asked:", user_text)
    
    agent_prompt = f"Answer this question and return the spoken answer as an audio file: {user_text}"
    agent_result = agent.run(agent_prompt)
    
    audio_path = extract_audio_path(agent_result)
    if audio_path and os.path.exists(audio_path):
        print("\nPlaying answer:")
        display(Audio(audio_path))
    else:
        print("\nNo audio file found. Agent result:")
        print(agent_result)
    
    if isinstance(agent_result, str):
        print("\nAgent text result:", agent_result)
    else:
        print("\nAgent result:", agent_result)

def detect_language(text):
    lang = langid.classify(text)[0]
    return 'ar' if lang == 'ar' else 'en'

def voice_query_and_speak(agent, duration=7):
    audio_file = record_audio(duration=duration)
    print("Transcribing...")
    with open(audio_file, "rb") as f:
        transcript = openai.Audio.transcribe("whisper-1",f)
        user_text = transcript["text"]

    print("\nYou asked:", user_text)
    agent_result = agent.run(user_text)

    # Convert answer to speech
    lang = detect_language(user_text)
    tts = gTTS(agent_result, lang=lang)
    tts_file = "agent_reply.mp3"
    tts.save(tts_file)

    print("\nPlaying answer:")
    display(Audio(tts_file))

def text_query_and_reply(agent):
    user_text = input("Your question: ")
    print(f"\nYou asked: {user_text}")

    agent_result = agent.run(user_text)
    print("\nAgent (text answer):", agent_result)

    # Convert answer to speech
    lang = detect_language(user_text)
    tts = gTTS(agent_result, lang=lang)
    tts_file = "agent_reply.mp3"
    tts.save(tts_file)

    print("\nPlaying answer:")
    display(Audio(tts_file))

def unified_convo_agent():
    print("🎙️ Welcome to RimalAI! Ask about Saudi culture, landmarks, or poetry.")
    while True:
        mode = input("Type 'v' for voice, 't' for text, or 'q' to quit: ").strip().lower()
        if mode == 'q':
            print("Goodbye!")
            break
        elif mode == 'v':
            voice_query_and_speak(agent)
        elif mode == 't':
            text_query_and_reply(agent)
        else:
            print("Invalid option. Please type 'v', 't', or 'q'.")
```

```python
# mapbox integration

# ================== Imports ==================
import requests
import wikipediaapi
import ast
import json
import re
from langchain.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.prompts import PromptTemplate
from langchain.chains import RetrievalQA
from langchain.chat_models import ChatOpenAI
from langchain.tools import tool
from langchain.agents import initialize_agent, AgentType
from langchain.agents import Tool as LangchainTool


# ================== 1. Load Vector Store ==================
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
landmark_vectordb = FAISS.load_local(
    folder_path="rimalai_faiss",
    embeddings=embedding_model,
    allow_dangerous_deserialization=True
)


# ================== 2. Define RetrievalQA Chain ==================
llm = ChatOpenAI(model_name="gpt-4o", temperature=0.3)

custom_prompt = PromptTemplate(
    input_variables=["context", "question"],
    template="""
    You are a helpful assistant. Answer the following question using the provided context.
    If the question is in Arabic, answer in Arabic. If in English, answer in English.
    Context:
    {context}
    Question: {question}
    Answer:
    """
)

qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=landmark_vectordb.as_retriever(),
    return_source_documents=True,
    chain_type_kwargs={"prompt": custom_prompt}
)

def query_landmark(query: str):
    response = qa_chain(query)
    answer = response["result"]
    media_results = []
    for doc in response["source_documents"]:
        media = doc.metadata.get("media", {})
        if isinstance(media, str):
            try:
                media = ast.literal_eval(media)
            except Exception:
                media = {}
        video_link = media.get("videos")[0] if media.get("videos") else None
        image_link = media.get("images")[0] if media.get("images") else None
        audio_link = media.get("audio")[0] if media.get("audio") else None
        media_results.append({
            "title": doc.metadata.get("name"),
            "image": image_link,
            "video": video_link,
            "audio": audio_link
        })
    return {"answer": answer, "media": media_results}


# ================== 3. Mapbox API ==================


def get_coordinates_from_mapbox(place_name):
    geocode_url = f"https://api.mapbox.com/geocoding/v5/mapbox.places/{place_name}.json"
    params = {
        "access_token": MAPBOX_API_KEY,
        "limit": 1,
        "country": "SA",  # Saudi Arabia
        "bbox": "34.5,16.0,55.7,32.2"
    }
    response = requests.get(geocode_url, params=params).json()
    features = response.get("features")
    if features:
        lon, lat = features[0]["center"]
        return lon, lat
    return None, None

def get_mapbox_image(place_name):
    lon, lat = get_coordinates_from_mapbox(place_name)
    if lon is None or lat is None:
        return f"https://api.mapbox.com/styles/v1/mapbox/streets-v11/static/0,0,1/600x600?access_token={MAPBOX_API_KEY}"
    url = f"https://api.mapbox.com/styles/v1/mapbox/streets-v11/static/{lon},{lat},5,0/600x600?access_token={MAPBOX_API_KEY}"
    return url


# ================== 4. Google Image Search ==================


def google_image_search(query):
    search_url = "https://www.googleapis.com/customsearch/v1"
    params = {
        "q": query,
        "cx": GOOGLE_CSE_ID,
        "key": GOOGLE_CSE_API_KEY,
        "searchType": "image",
        "num": 3
    }
    response = requests.get(search_url, params=params).json()
    return [item["link"] for item in response.get("items", [])]


# ================== 5. Wikipedia Search ==================
def search_wikipedia(query, lang="en"):
    wikipedia.set_lang(lang)
    try:
        page = wikipedia.page(query)
        return page.summary, page.images[:3]
    except Exception:
        return None, []


# ================== 6. Main Landmark Agent ==================
def landmark_agent(query):
    result = query_landmark(query)
    answer = result.get("answer")
    media = result.get("media")

    if answer and len(answer) > 20:  # Enough content
        return {
            "source": "vector_search",
            "title": query,
            "info": answer,
            "media": media,
            "map_url": get_mapbox_image(query)
        }
    else:
        # Fallback to Wikipedia + Google Images
        summary_en, wiki_images = search_wikipedia(query, lang="en")
        summary_ar, _ = search_wikipedia(query, lang="ar")
        google_images = google_image_search(query)
        return {
            "source": "wikipedia",
            "title": query,
            "info_en": summary_en,
            "info_ar": summary_ar,
            "images": google_images + wiki_images,
            "map_url": get_mapbox_image(query)
        }


# ================== 7. Tool Definition ==================
@tool
def landmark_search_tool(query: str) -> dict:
    """Search for information and media about Saudi landmarks. Accepts Arabic or English queries."""
    return landmark_agent(query)


# ================== 8. Initialize Router Agent ==================
llm = ChatOpenAI(model_name="gpt-4o", temperature=0)

landmark_tool = LangchainTool(
    name="landmark_search",
    func=landmark_search_tool,
    description="Useful for searching Saudi landmarks and tourist places. Accepts questions like 'Tell me about Al-Ula' or 'ما هي معالم الرياض؟'"
)

agent = initialize_agent(
    tools=[landmark_tool],
    llm=llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True,
    handle_parsing_errors=True
)


# ================== 9. Extract Media & Map from Agent Response ==================
def extract_media_and_map(agent_response):
    try:
        # Find observation
        observation_start = agent_response.find("Observation: ")
        if observation_start == -1:
            # Fallback: look for final answer
            final_answer_start = agent_response.find("Final Answer: ")
            if final_answer_start == -1:
                return {"media": {"images": [], "videos": [], "audio": []}, "map_url": None}

            final_answer = agent_response[final_answer_start + len("Final Answer: "):]
            final_answer_end = final_answer.find("\n> Finished chain.")
            if final_answer_end != -1:
                final_answer = final_answer[:final_answer_end]

            image_urls = re.findall(r'\[.*?\]\((https://.*?\.png)\)', final_answer)
            map_urls = re.findall(r'\[.*?\]\((https://.*?mapbox.*?)\)', final_answer)

            return {
                "media": {
                    "images": image_urls,
                    "videos": [],
                    "audio": []
                },
                "map_url": map_urls[0] if map_urls else None
            }

        observation_text = agent_response[observation_start + len("Observation: "):]
        observation_end = observation_text.find("\nThought:")
        if observation_end != -1:
            observation_text = observation_text[:observation_end]

        observation = json.loads(observation_text)

        # Extract media
        media = observation.get('media', [])
        extracted_media = {"images": [], "videos": [], "audio": []}

        for item in media:
            if item.get('image'):
                extracted_media["images"].append(item['image'])
            if item.get('video'):
                extracted_media["videos"].append(item['video'])
            if item.get('audio'):
                extracted_media["audio"].append(item['audio'])

        map_url = observation.get('map_url')

        return {
            "media": extracted_media,
            "map_url": map_url
        }

    except Exception as e:
        print(f"Error extracting media and map: {str(e)}")
        return {"media": {"images": [], "videos": [], "audio": []}, "map_url": None}

```

```python
import os
import openai
import langid
import tempfile
import ast
import re
import os
import speech_recognition as sr
from gtts import gTTS
from IPython.display import Audio, display
from langchain.chat_models import ChatOpenAI
from langchain.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate
from langchain.tools import Tool
from langchain.agents import initialize_agent, AgentType

# intent classification based on user query
def classify_user_intent(user_query: str) -> str:
    """
    Uses GPT-4o to classify user intent: text, poem, image, voice.
    """
    system_msg = (
        "You are an AI assistant that classifies user questions into one of the following intents: "
        "'text', 'poem', 'image', or 'voice'. "
        "Return ONLY one word without explanation."
    )
    user_msg = f"User asked: {user_query}\nWhat is the intent?"
    response = client.ChatCompletion.create(
        model="gpt-4o",
        messages=[
            {"role": "system", "content": system_msg},
            {"role": "user", "content": user_msg}
        ],
        max_tokens=1
    )
    return response.choices[0].message.content.strip().lower()


# === 1. Load saved FAISS ===
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vectordb = FAISS.load_local(
    folder_path="rimalai_faiss",
    embeddings=embedding_model,
    allow_dangerous_deserialization=True
)

# === 2. Initialize LLM ===
openai.api_key = os.getenv("OPENAI_API_KEY")
client = openai
llm = ChatOpenAI(model_name="gpt-4o", temperature=0.3)

# === 3. Define prompt templates ===
custom_prompt = PromptTemplate(
    input_variables=["context", "question"],
    template="""
    You are a helpful assistant. Answer the following question using the provided context.
    If the question is in Arabic, answer in Arabic. If in English, answer in English.

    Context:
    {context}

    Question: {question}
    Answer:
    """
)

# === 4. QA Chains setup ===
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=vectordb.as_retriever(),
    return_source_documents=True,
    chain_type_kwargs={"prompt": custom_prompt}
)

def query_with_media(query: str):
    response = qa_chain(query)
    answer = response["result"]
    media_results = []
    for doc in response["source_documents"]:
        media = doc.metadata.get("media", {})
        if isinstance(media, str):
            try:
                media = ast.literal_eval(media)
            except Exception:
                media = {}
        video_link = media.get("videos")[0] if media.get("videos") else None
        image_link = media.get("images")[0] if media.get("images") else None
        audio_link = media.get("audio")[0] if media.get("audio") else None
        media_results.append({
            "title": doc.metadata.get("name"),
            "image": image_link,
            "video": video_link,
            "audio": audio_link
        })
    return {"answer": answer, "media": media_results}

# === Tools Functions ===

# @tool 1: QA with Media
db_tool = Tool(
    name="SaudiRimalRetrievalQAWithMedia",
    func=query_with_media,
    description="Answer questions about Saudi landmarks, Vision 2030, and culture, returning answer and media links."
)

# @tool 2: Arabic Culture Poem + Image
ARABIC_CULTURE_KEYWORDS = [
    "arab", "saudi", "bedouin", "najd", "hijaz", "mecca", "medina", "riyadh", "alula", "al-ula",
    "najdi", "ghazal", "nabati", "jeddah", "diriyah", "desert", "camels", "dates", "gahwa", "coffee",
    "palms", "arabian", "oasis", "sand", "tribe", "poet", "poetry", "heritage", "tradition",
    "قصيدة", "شعر", "عربي", "سعودي", "بدوي", "نجد", "الحجاز", "مكة", "المدينة", "الرياض", "العلا",
    "جدة", "الدرعية", "صحراء", "جمل", "تمر", "قهوة", "نخيل", "قبيلة", "شاعر", "تراث", "تقاليد"
]

def is_arabic_culture(prompt):
    prompt_lower = prompt.lower()
    return any(kw in prompt_lower for kw in ARABIC_CULTURE_KEYWORDS)

def generate_arabic_poem(prompt):
    system_prompt = (
        "أنت شاعر عربي محترف. "
        "اكتب قصيدة عربية أصيلة حول الموضوع الذي يطلبه المستخدم، بأسلوب تقليدي (مثل النبطي أو الغزل أو الحكمة). "
        "بعد القصيدة، قدم ترجمة إنجليزية دقيقة لها تحت عنوان 'Translation:'."
    )
    user_prompt = (
        f"اكتب قصيدة عربية حول: {prompt}\n"
        "ثم ترجم القصيدة إلى الإنجليزية تحت عنوان 'Translation:'."
    )
    response = client.ChatCompletion.create(
        model="gpt-4o",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ],
        max_tokens=400
    )
    return response.choices[0].message.content.strip()

def generate_image(prompt, n=1, size="1024x1024"):
    response = openai.Image.create(
        model="dall-e-3",
        prompt=prompt,
        n=n,
        size=size
    )
    return [img["url"] for img in response["data"]]

def arabic_culture_poem_and_image(prompt):
    if not is_arabic_culture(prompt):
        return {
            "message": (
                "⚠️ This tool is dedicated to Arabic culture and heritage. "
                "Please provide a prompt related to Arabic themes, history, or traditions."
            )
        }
    poem = generate_arabic_poem(prompt)
    images = generate_image(prompt)
    return {
        "poem": poem,
        "image_urls": images
    }

culture_tool = Tool(
    name="ArabicCulturePoemAndImage",
    func=arabic_culture_poem_and_image,
    description=(
        "Given a prompt related to Arabic culture, generates an original Arabic poem (with English translation) and a DALL-E image. "
        "If the prompt is not about Arabic culture, returns a polite message."
    )
)

# @tool 3: Text-to-Speech
def tts_tool_func(text: str) -> str:
    lang = langid.classify(text)[0]
    tts = gTTS(text, lang=lang if lang in ['en', 'ar'] else 'en')
    with tempfile.NamedTemporaryFile(delete=False, suffix=".mp3") as fp:
        tts.save(fp.name)
        return fp.name

tts_tool = Tool(
    name="GoogleTextToSpeech",
    func=tts_tool_func,
    description="Convert an answer in Arabic or English to speech (returns MP3 file path)."
)

# Voice Input Function (not a @tool)
def record_and_transcribe(duration=7, filename="user_question.wav"):
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        audio = r.record(source, duration=duration)
        with open(filename, "wb") as f:
            f.write(audio.get_wav_data())
    with open(filename, "rb") as f:
             transcript = openai.Audio.transcribe("whisper-1", f)
    user_text = transcript['text']
    return transcript["text"]

# === 6. Agent Setup (Extended with New Tools) ===

# Your existing tools
tools = [db_tool, culture_tool, tts_tool]

# Add new tools
tools += [landmark_search_tool, search_poetry_tool]

# Reinitialize agent with all tools
agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent=AgentType.OPENAI_FUNCTIONS,
    verbose=True
)


# === Voice Input Workflow ===
def record_audio(duration=7, filename="user_question.wav"):
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        audio = r.record(source, duration=duration)
        with open(filename, "wb") as f:
            f.write(audio.get_wav_data())
    return filename

def extract_audio_path(agent_result):
    """Extract .mp3 file path from agent result string."""
    match = re.search(r'([/\w\-.]+\.mp3)', agent_result)
    if match:
        path = match.group(1)
        if path.startswith("sandbox:"):
            path = path.replace("sandbox:", "")
        if not os.path.exists(path) and os.path.exists(os.path.basename(path)):
            path = os.path.basename(path)
        return path
    return None

def voice_query_agent_workflow(agent, duration=7):
    audio_file = record_audio(duration=duration)
    print("Transcribing...")
    with open(audio_file, "rb") as f:
        transcript = openai.Audio.transcribe("whisper-1", f)
    user_text = transcript['text']
    print("\nYou asked:", user_text)
    
    agent_prompt = f"Answer this question and return the spoken answer as an audio file: {user_text}"
    agent_result = agent.run(agent_prompt)
    
    audio_path = extract_audio_path(agent_result)
    if audio_path and os.path.exists(audio_path):
        print("\nPlaying answer:")
        display(Audio(audio_path))
    else:
        print("\nNo audio file found. Agent result:")
        print(agent_result)
    
    if isinstance(agent_result, str):
        print("\nAgent text result:", agent_result)
    else:
        print("\nAgent result:", agent_result)

def detect_language(text):
    lang = langid.classify(text)[0]
    return 'ar' if lang == 'ar' else 'en'

def voice_query_and_speak(agent, duration=7):
    audio_file = record_audio(duration=duration)
    print("Transcribing...")
    with open(audio_file, "rb") as f:
        transcript = openai.Audio.transcript("whisper-1", f)
    
    user_text = ["text"]
    print("\nYou asked:", user_text)

    agent_result = agent.run(user_text)

    # Convert answer to speech
    lang = detect_language(user_text)
    tts = gTTS(agent_result, lang=lang)
    tts_file = "agent_reply.mp3"
    tts.save(tts_file)

    print("\nPlaying answer:")
    display(Audio(tts_file))

def text_query_and_reply(agent):
    user_text = input("Your question: ")
    print(f"\nYou asked: {user_text}")

    agent_result = agent.run(user_text)
    print("\nAgent (text answer):", agent_result)

    # Convert answer to speech
    lang = detect_language(user_text)
    tts = gTTS(agent_result, lang=lang)
    tts_file = "agent_reply.mp3"
    tts.save(tts_file)

    print("\nPlaying answer:")
    display(Audio(tts_file))

#smart conversational loop with GPT-based intent detection
def unified_convo_agent_pro():
    print("🌟 Welcome to RimalAI Pro! Ask about Saudi culture, Vision 2030, poetry, images, or request voice answers.")
    while True:
        mode = input("\nType 'v' for voice, 't' for text, or 'q' to quit: ").strip().lower()
        if mode == 'q':
            print("👋 Goodbye!")
            break
        if mode == 'v':
            user_query = record_and_transcribe()
            print(f"\n🗣️ You asked (via voice): {user_query}")

            # Voice mode always replies with voice
            agent_result = agent.run(user_query)
            print("\n📝 Agent:", agent_result)
            audio_path = tts_tool_func(agent_result)
            print("\n🔊 Playing answer...")
            display(Audio(audio_path))

        elif mode == 't':
            user_query = input("Your question: ")
            print(f"\n📝 You asked (via text): {user_query}")

            
            intent = classify_user_intent(user_query)
            print(f"🤖 Detected intent: {intent}")

            if intent == "poem":
                result = arabic_culture_poem_and_image(user_query)
                print("\n📝 Poem:\n", result.get("poem", ""))
                if "image_urls" in result:
                    print("\n🖼️ Image(s):")
                    for url in result["image_urls"]:
                        print(url)

            elif intent == "image":
                images = generate_image(user_query)
                print("\n🖼️ Generated image(s):")
                for url in images:
                    print(url)

            elif intent == "voice":
                agent_result = agent.run(user_query)
                print("\n📝 Agent:", agent_result)
                audio_path = tts_tool_func(agent_result)
                print("\n🔊 Playing answer...")
                display(Audio(audio_path))

            else:  # Default → text
                agent_result = agent.run(user_query)
                print("\n📝 Agent:", agent_result)

        else:
            print("⚠️ Invalid option. Please type 'v', 't', or 'q'.")
```

```python
def evaluate_and_diagnose(query, expected_ids, vectordb, k=3):
    # Step 1: Retrieve top-k documents
    results = vectordb.similarity_search(query, k=k)
    retrieved_ids = [doc.metadata.get("id", None) for doc in results]

    # Step 2: Match IDs
    found_ids = [doc_id for doc_id in expected_ids if doc_id in retrieved_ids]

    # Step 3: Compute Metrics
    precision = len(found_ids) / k
    recall = len(found_ids) / len(expected_ids) if expected_ids else 0
    mrr = sum([1 / (retrieved_ids.index(doc_id) + 1) for doc_id in found_ids]) / len(found_ids) if found_ids else 0

    # ✅ Step 4: Print formatted output
    print(f"\n🔍 Query: {query}")
    print(f"📊 Precision@{k}: {precision:.2f}")
    print(f"📍 Recall@{k}: {recall:.2f}")
    print(f"📈 MRR: {mrr:.2f}")
    print("-" * 40)  # فاصل بين النتائج

    return precision, recall, mrr
```

```python
from bert_score import score

def evaluate_bertscore(reference, generated):
    """
    Evaluate the semantic similarity between reference and generated texts using BERTScore.
    
    Parameters:
    - reference: str, the ground truth answer
    - generated: str, the model's generated answer

    Returns:
    - dict: scores for precision, recall, and F1
    """
    P, R, F1 = score([generated], [reference], lang="ar", rescale_with_baseline=True)
    return {
        "precision": round(P.item(), 3),
        "recall": round(R.item(), 3),
        "f1": round(F1.item(), 3)
    }
```

```python
from langsmith.evaluation import EvaluationResult, run_evaluator
from sentence_transformers import SentenceTransformer, util

# Load the semantic similarity model once
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

@run_evaluator
def qa_accuracy(run, example) -> EvaluationResult:
    predicted = run.outputs.get("output", run.outputs.get("result", ""))
    reference = example.outputs.get("answer", "")

    # Check for empty strings to avoid errors
    if not reference.strip() or not predicted.strip():
        score = 0.0
        similarity = 0.0
    else:
        # Extract embeddings for both texts
        embeddings1 = model.encode(reference, convert_to_tensor=True)
        embeddings2 = model.encode(predicted, convert_to_tensor=True)

        # Compute cosine similarity
        similarity = util.pytorch_cos_sim(embeddings1, embeddings2).item()

        # Determine success score
        score = 1.0 if similarity > 0.75 else 0.0  # Adjust threshold as needed

    # Optional: print similarity for analysis
    print(f"Similarity: {similarity:.2f} | Score: {score} | Reference: {reference} | Predicted: {predicted}")

    return EvaluationResult(key="qa_accuracy", score=score)
```
